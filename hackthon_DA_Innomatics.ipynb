7. 38.7 & 61.2 is the percentage of getting bat and field in feature 'toss_descision'

```python
field=0
bat=0
for x,y in matc.iterrows():
    if y["toss_decision"]=="field":
        if y["toss_winner"]== y["winner"]:
            field+=1
    else:
        if y["toss_winner"]== y["winner"]:
            
            bat+=1
```
